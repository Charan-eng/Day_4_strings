Palidrome

```python
str_16=input("enter the name to check palidrome")
if str_16==str_16[::-1].title() or str_16==str_16[::-1].lower() or str-16==str_16[::-1].upper():
    print("the name is palindrome")
else:
     print("the name is not palindrome")
```

string concatenation

1) % operator method

```python
name="charan"
```

```python
print("hello my name is",name)
```

```python
salary=100000
prof="developer"
```

```python
print("my name is %s and salary is %d and my prof is %s"%(name,salary,prof))
```

2) str.formate method

```python
"hello, {} My salary is {} and my prof is {}!".format(name,salary,prof)
```

3) f-string method

```python
f"hello my name is {name}, my salary is {salary}, my prof is {prof}"
```
